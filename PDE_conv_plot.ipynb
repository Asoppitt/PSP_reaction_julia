```julia
using Plots, LaTeXStrings, Printf
```

```julia
k = 0.938
C_0 = 2.1
omega_mean=10.0
B=(1.5*C_0)
bc_k=0.3
D=2*C_0*k/((B^2)*omega_mean)
height_domain =0.01
length_domain =0.20
dt=0.001
nt=200
T=dt*nt
nknots_array=2 .^(2:7)
error_num=zeros(length(nknots_array))
read!("conv_study_error", error_num);
```

```julia
plt=plot(nknots_array,error_num,
label="",
ylabel=L"\Vert u_{n_{mesh}}-u_{1024} \Vert_{L^2}",
xlabel="N Spatial Grid Points")
display(plt)


err_ticks_log = floor(minimum(log10.(error_num))):ceil(maximum(log10.(error_num)))
plt2=plot(nknots_array,log10.(error_num),
label="",
ylabel=L"\Vert u_{N}-u_{1024} \Vert_{L^2}",
xlabel=L"n_{mesh}",
ylims=[err_ticks_log[1],maximum(log10.(error_num))])
err_ticks = vcat([(1:10).*(10^err_ticks_log[i]) for i in 1:(length(err_ticks_log)-1)]...)
err_ticks = err_ticks[err_ticks.<=maximum(error_num)]
err_ticks_string = [i%10==1 ? @sprintf("%1.2e",err_ticks[i]) : "" for i in 1:length(err_ticks)]
yticks!(plt2, (log10.(err_ticks), err_ticks_string))
display(plt2)
savefig(plt2, "MoL_error_log")
```
